```python
img_array= os.listdir()
img_array[0]
```

```python
l = pd.DataFrame(columns=range(1,40001))
for i,k in zip(img_array,range(len(img_array))):
    image_path = path1+'\\'+i
    image = cv2.imread(image_path)
    img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    img = cv2.resize(img,(200,200),interpolation=cv2.INTER_AREA)
    
    mg = img.reshape(1,40000)
    l.loc[k] = mg[0]
```
